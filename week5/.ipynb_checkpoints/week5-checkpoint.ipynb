## Cost function of Neural Network

\begin{gather*} J(\Theta) = - \frac{1}{m} \sum_{i=1}^m \sum_{k=1}^K \left[y^{(i)}_k \log ((h_\Theta (x^{(i)}))_k) + (1 - y^{(i)}_k)\log (1 - (h_\Theta(x^{(i)}))_k)\right] + \frac{\lambda}{2m}\sum_{l=1}^{L-1} \sum_{i=1}^{s_l} \sum_{j=1}^{s_{l+1}} ( \Theta_{j,i}^{(l)})^2\end{gather*}

with 
1. L = total number of layers
2. Sl = number of units (without bias unit) in input layer
3. k = number of output units
4. h_theta(x)_k = the hypothesis results in the kth output

## Backpropagation Algorithm to minimize cost function

we use backpropagation algorithm for to minimize the cost function. 

Intuition: \begin{gather*} Î´^{(l)}_j = "error" of node j in layer l \end{gather*}

### Gradient Checking

a numeric way to comput gradient:
(J(theta + epsilon) - J(theta - epsilon)) / 2 * epsilon

we can compare the numeric result of gradient and the backpro result to see how close they are

Note: before we start the training process with the network, turn off gradient checking because it is not efficient.

### Random Initialization

we cant use all 0 thetas to start the minimazition process because after each gradient update the thetas will also be same. 

So we set theta to random numbers to break the symmetry setting.

## The whole process

- Design the layout of the neural network:
    - Number of input units = dimension of features $ x^{(i)}\$
    - Number of output units = number of classes
    - Number of hidden units per layer = tradeout between accuracy and computing effect
    - Defaulfs: 1 hidden layer. if more than one then better with same units number in each hidden layers

- Train the network
    1. Randomly initialize the thetas (weights/parameters)
    2. Implement forward progagation to obtain $ h_{\Theta}(x^{(i)}) \$ for any x
    3. Implement the cost function
    4. implement backpro to get partial derivatives (gradient)
    5. Use gradient check, then disable it
    6. use gradient descent(for loop) or optimization function to minimiza the cost functio in order to get theta
