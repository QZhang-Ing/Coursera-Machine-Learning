## Non-linear Classification
 By non-linear Classification problem, we have to use high grade polynomial to fit our training datasets. 
 In case of a quadratic polynomial, (x_1)2, (x_1 * x_2) ... the algorithms has a O(n2) complexity . When a cubic polynomial, it goes up to O(n3).

Hence, it is not a good way to train non-linear problem

a1 a2 a3 are neurons. Output is our hypothesis. Layer 1 is input layer. layer 3 is output layer. layer 2 is hidden layer which can not be observed in supervised learning.

\begin{align*}& a_i^{(j)} = \text{"activation" of unit $i$ in layer $j$} \newline& \Theta^{(j)} = \text{matrix of weights controlling function mapping from layer $j$ to layer $j+1$}\end{align*}

If we had one hidden layer, it would look like:

\begin{align*} a_1^{(2)} = g(\Theta_{10}^{(1)}x_0 + \Theta_{11}^{(1)}x_1 + \Theta_{12}^{(1)}x_2 + \Theta_{13}^{(1)}x_3) \newline a_2^{(2)} = g(\Theta_{20}^{(1)}x_0 + \Theta_{21}^{(1)}x_1 + \Theta_{22}^{(1)}x_2 + \Theta_{23}^{(1)}x_3) \newline a_3^{(2)} = g(\Theta_{30}^{(1)}x_0 + \Theta_{31}^{(1)}x_1 + \Theta_{32}^{(1)}x_2 + \Theta_{33}^{(1)}x_3) \newline h_\Theta(x) = a_1^{(3)} = g(\Theta_{10}^{(2)}a_0^{(2)} + \Theta_{11}^{(2)}a_1^{(2)} + \Theta_{12}^{(2)}a_2^{(2)} + \Theta_{13}^{(2)}a_3^{(2)}) \newline \end{align*}

If network has sj units in layer j and sj+1 units in layer j+1, then Θ(j) will be of dimension s j+1 * (sj+1).

a layer (layer 2) is actually a weigted process. a1 = g(z1). What a layer does is weighting input values and then pas to the g(z) funciton.
