## What is Machine Learning

According to Tom Mitchell(1998):
Well-posedLearning Problem: A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E.

## Classification of Machine Learning Algorithms:

1. Supervised learning
2. Unsupervised learning
3. Reinforcement learning
4. etc.

### Supervised Learning
key word: **"Right answer" given**

Two kind of problem:
1. Regression: predict **continuous** values, such as housing price prediction
2. Classification: predict **discrete** values, such as Cancer Type(malignant or benign) prediction

### Unsupervised Learning
key word: **Predict Pattern with no sign of what is right what is wrong**

e.g. news searching result grupping problem; Create merket segments based on customer data; 

## What is Model?

A supervised learning problem can be described as:
given a training set, to learn a function h : X → Y so that h(x) is a “good” predictor for the corresponding value of y. For historical reasons, this function h is called a hypothesis. This is the model representation of a machine learning problem. 
When y is continuous, it is a regression problem; otherwise, it is a classification problem.

## What is Cost Function?

A cost function is used to measure the accuracy of our hypothesis function h(x). 

                J = 1/2 * mean of squares of (h(x) - y)

also called **square error function** or **mean square error**

This '1/2' is a term for the computation of the gradient descent of the cost function: the derivative of the square error will generate '2', which will cancel out '1/2'

Ultimate goal is to get a line go cross all the possible points('right answers given'). If a line goes across all possible points, then J is 0. 

## How to get to the best hypothesis/minimize the cost function?

A graphic representation of cost function:


```python
from IPython.display import Image
Image("img/costfunction.png")
```

our goal is to play with theta and reach the bottom of the graph. 
By looking at the slope(derivative) of the J at a certain point, we get in which direction the value of the function will develop. Increase or decrease. If we follow the direction of the steepest descent, we will reach the bottom of the graph fastest. The size of each step is **alpha  learning rate**.

A smaller α would result in a smaller step and a larger α results in a larger step. The direction in which the step is taken is determined by the partial derivative of J(θ0,θ1). Depending on where one starts on the graph, one could end up at different points. The image above shows us two different starting points that end up in two different places.

Reach convergence, means J  = J - alpha * partial derivative of J, meaning the derivative == 0.

At each iteration j, one should simultaneously update the parameters θ1,θ2,...,θn

With this approach, we could manage to repeat until convergence regardless the sign of the slope for the derivative, because when slope is negative, theta increase. otherwise, theta decrease. However, we should pay attention to the learning rate alpha. When alpha to small, too slow; otherwise, may not converge.
