### How to make decision about 1 or 0?

So we notice that when z >= 0, g(z) >= 0.5 -> h(x) -> 1, otherwise, 0. 
And note that, z = thetaTx

By using z >= 0 or z < 0 we can find the **decision boundary**

## How to fit the parameters (theta) of Logistic function

We use cost function to find the fits for our parameters of regresion problems. However, because of the existing of g(z), the cost function introduced in the previous charpter is no more practical for logistic regression. It turns the cost function into a non convex function, which means it has multiple local minimums. In this case it may converges at a local minimus but not the global minimum.

## What is Overfitting

Such as using too higher a poly function (too many feartures), it may cause overfit / high varience. it is not a good representation of our problem because it not not a general solution.

### Addressing Overfitting

1. reduce number of features
    manually select or model select
2. regularization
    keeps all features but reduce the magnitute of parameters

### How regularization works?
